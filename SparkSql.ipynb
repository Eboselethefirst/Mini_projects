```python
import findspark as fs
fs.init()

from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
```

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/mpg.csv
```

```python
df = spark.read.csv("mpg.csv",header=True,inferSchema=True)
```

```python
df.printSchema()
```

```python
df.createTempView("mpg")
```

```python
spark.sql("select distinct(Origin) from mpg ").show()
```

```python
spark.sql("Select count(*) from mpg where Origin = 'Japanese'").show()
```

```python
df.show(7)
```

```python
spark.sql("select count(*) from mpg where MPG > 40").show()
```

```python
spark.sql("select year,count(*) as number_of_cars from mpg group by Year order by Year").show()
```

```python
spark.sql("Select max(MPG) from mpg").show()
```

```python
spark.stop()
```
